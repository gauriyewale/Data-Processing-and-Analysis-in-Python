```python
#1. Given customer bank saving data customer-savings.txt,please 
# answer the following questions using Python programming.
# What’s the average balance for male and female?
# What’s the average balance for while collar and blue collar in England?

#Open customer savings text file and use readlines function to save it into a list
fh = open('customer-savings.txt')
customer_savings = fh.readlines()

#Iterate over customer savings, strip the data and split it into substrings
for entry in range(0, len(customer_savings)):
	customer_savings[entry] = customer_savings[entry].strip().split(',')

#Initialize variables to store counts and sums
sum_bal_male = 0
sum_bal_female = 0
count_male = 0
count_female = 0
sum_bal_white_collar = 0
sum_bal_blue_collar = 0
count_white_collar = 0
count_blue_collar = 0

#Iterate over the list, check conditions, update counters and aggregate the balances
for i in range(0,len(customer_savings)):
	if customer_savings[i][3] == 'Female':
		count_female += 1
		sum_bal_female += float(customer_savings[i][-1])
	elif customer_savings[i][3] == 'Male':
		count_male += 1
		sum_bal_male += float(customer_savings[i][-1])

	if customer_savings[i][6] == 'White Collar' and customer_savings[i][5] == 'England':
		count_white_collar += 1
		sum_bal_white_collar += float(customer_savings[i][-1])
	elif customer_savings[i][6] == 'Blue Collar' and customer_savings[i][5] == 'England':
		count_blue_collar += 1
		sum_bal_blue_collar += float(customer_savings[i][-1])	

#Calculate averages
average_bal_male = sum_bal_male / count_male
average_bal_female = sum_bal_female / count_female
average_bal_white_collar = sum_bal_white_collar / count_white_collar
average_bal_blue_collar = sum_bal_blue_collar / count_blue_collar

#Print average balances
print(f"Average balance of males: {average_bal_male}")
print(f"Average balance of females: {average_bal_female}")
print(f"Average balance of white collar in England: {average_bal_white_collar}")
print(f"Average balance of blue collar in England: {average_bal_blue_collar}")
```

```python
#2. Combine multiple files
#Use Python programming to combine two text files: customer-status.txt and sales.txt


#Read both files
fh_status = open('customer-status.csv')
fh_sales = open('sales.csv')

#Store the files in the form of lists
customer_status = fh_status.readlines()
customer_sales = fh_sales.readlines()

#Initiate dictionaries
merged = {}
cust_status = {}
cust_sales = {}

#Iterate over customer_sales
for entry in range(0, len(customer_sales)):
	#Strip the data and split it into substrings
	customer_sales[entry] = customer_sales[entry].strip().split(',')
	#Save the data in the form of a dictionary with account number as key
	cust_sales[customer_sales[entry][0]] = customer_sales[entry][1:]

#Iterate over customer_status
for entry in range(0, len(customer_status)):
	#Strip the data and split it into substrings
	customer_status[entry] = customer_status[entry].strip().split(',')
	#Save the data in the form of a dictionary with account number as key
	cust_status[customer_status[entry][0]] = customer_status[entry][2:]

#Delete the 1st record from both dictionaries as it's the header
del cust_status['account number']
del cust_sales['account number']

#Loop over the customer sales and status dictionaries
for key in cust_sales.keys():
	 for k in cust_status.keys():
	 	#Check if the keys match
	 	if key == k:
	 		#Add the status column to the sales dictionary and convert 
	 		#the entire thing to a string
	 		merged[key] = ', '.join(cust_sales[key]) + ', ' + cust_status[key][-1]

#Create a csv file and save the dictionary 
with open('3.2_lab/merge.csv', 'w') as f:
    for key in merged.keys():
        f.write("%s,%s\n"%(key,merged[key]))
```
